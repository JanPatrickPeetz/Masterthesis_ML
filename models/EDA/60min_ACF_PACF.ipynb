```python
import pandas as pd
import matplotlib.pyplot as plt
import pyarrow.parquet as pq
import seaborn as sns
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import statsmodels.api as sm
```

```python
df = pd.read_parquet('/Users/paddy/Documents/GitHub/Masterthesis_ML/data/02_timeseries_zone70.parquet')

# # Convert the date column to datetime
df['date'] = pd.to_datetime(df['tpep_pickup_datetime'])

# set date as index
df_1day = df.set_index('date')
```

```python
""" # making intervals of 1 day
df_1day = (pd.to_datetime(df['tpep_pickup_datetime'], unit='min')
    .dt.floor('60min') 
    .value_counts()
    .rename_axis('date')
    .reset_index(name='rides')
) """

df_1day = df_1day.resample('H').size()

# sort by date
#df_1day = df_1day.sort_values(by='date')

```

```python
# Recreate the ACF and PACF plots with royal blue color for lines and dots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# ACF plot with royal blue color
sm.graphics.tsa.plot_acf(df_1day, lags=48, ax=ax1, color='royalblue')
ax1.set_title('Hourly Autocorrelation Function (ACF)')

# PACF plot with royal blue color
sm.graphics.tsa.plot_pacf(df_1day, lags=48, ax=ax2, color='royalblue')
ax2.set_title('Hourly Partial Autocorrelation Function (PACF)')

plt.tight_layout()

#safe plot
plt.savefig('/Users/paddy/Documents/GitHub/Masterthesis_ML/plots/02_acf_pacf_60min.png')

plt.show()
```
